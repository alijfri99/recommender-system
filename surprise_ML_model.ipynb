```python
import pandas as pd
from surprise import Dataset
from surprise import Reader

from surprise import Reader, Dataset, SVD
from surprise.model_selection.validation import cross_validate
```

```python
ratings = pd.read_csv("./ir_assignment_table_ratings.csv") 

ratings.head()
```

```python
reader = Reader()
data = Dataset.load_from_df(ratings[['User-ID', 'id', 'Book-Rating']], reader)
svd = SVD()
```

```python
# Run 5-fold cross-validation and print results
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
trainset = data.build_full_trainset()
svd.fit(trainset)
```

```python
df_title = pd.read_csv('./ir_assignment_table_books.csv', encoding = "utf-8" )
df_title.head(10)
```

```python
df_title = df_title[["id","ISBN" , "Book-Title" , "Book-Author"]]
df_title.head()
```

```python
titles = df_title.copy()
titles['Estimate_Score'] = titles['id'].apply(lambda x: svd.predict(989, x).est)
```

```python
titles = titles.sort_values(by=['Estimate_Score'], ascending=False)
titles.head(10)
```
